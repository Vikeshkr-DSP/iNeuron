```python
n=int(input('Enter a number: '))

for i in range(n):
    for j in range (i+1):
        print('*',end='')
    print('')

for i in range(n-1,0,-1):
    for j in range(i):
        print('*',end='')
    print('')
```

Q2. Write a Python program to reverse a word after accepting the input from the user.

```python
wrd=input('Enter any random word: ')
print(wrd,'in reverse order:',wrd[::-1])
```
